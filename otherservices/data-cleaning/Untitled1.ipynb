```python
columns=['ID', 'exerciseCategory', 'exerciseName', 'level', 'type', 'direction', 'modality',
       'joint']
originalFile=pd.read_excel("resist_train_planner.xls",header=18,names=columns)
```

```python
jsonArray=[]
for row in range(originalFile.exerciseCategory.count()):
    dictionary={}
    for i in columns:
        dictionary[i]=originalFile[i][row]
    dictionary["targetMuscleName"]=[]
    photosUrl=[]
    mainPhoto=""
    dictionary["photos"]={}
    dictionary["photos"]["mainPhoto"]=mainPhoto
    dictionary["photos"]["photosUrl"]=photosUrl
    jsonArray.append(dictionary)
```
